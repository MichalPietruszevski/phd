```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel("/content/drive/MyDrive/DOKTORAT/ANALIZY/DANE/baza_danych.xlsx")
df_crp = pd.read_excel("/content/drive/MyDrive/DOKTORAT/ANALIZY/DANE/OPUS 21 hsCRP, sIgA - WYNIKI DO BAZY.xlsx")
```

```python
df_crp = df_crp.iloc[2:,:]
df_crp = df_crp[[i for i in df_crp.columns if "Unnamed" not in i]]
```

```python
df_crp['id'] = df_crp["hsCRP"]
df_crp['hsCRP'] = df_crp['NR KATALOGOWY BXE0617A']
df_crp = df_crp.drop(columns=['NR KATALOGOWY BXE0617A'])
df_crp['sIgA'] = df_crp['sIgA NR CAT 1-1602 SALIMMETRICS']
df_crp = df_crp.drop(columns=['sIgA NR CAT 1-1602 SALIMMETRICS'])
```

```python
df = df.merge(df_crp, on='id', how='left')
```
